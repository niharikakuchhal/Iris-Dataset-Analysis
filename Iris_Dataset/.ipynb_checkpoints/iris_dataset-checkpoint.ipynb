<h3 style='color:red'>Removing Outliers</h3>

```python
# IQR
Q1 = np.percentile(iris['SepalWidthCm'], 25,method = 'midpoint')

Q3 = np.percentile(iris['SepalWidthCm'], 75,method = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", iris.shape)

# Upper bound
upper = np.where(iris['SepalWidthCm'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(iris['SepalWidthCm'] <= (Q1-1.5*IQR))

# Removing the Outliers
iris.drop(upper[0], inplace = True)
iris.drop(lower[0], inplace = True)

print("New Shape: ", iris.shape)

sns.boxplot(x='SepalWidthCm', data=iris)
```

```python
X = iris.drop('Species',axis=1) #label
y = iris.Species  # target
```

##### Split dataset into training and testing dataset 

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8, random_state=121)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

### Machine Learning Model
#### Random Forest Classification

```python
clf = RandomForestClassifier(random_state=121)
```

```python
clf.fit(X_train, y_train)
```

```python
prediction = clf.predict(X_test)
```

### Evaluation of model

```python
acc = accuracy_score(y_test, prediction)
```

```python
print('Accuracy: ',acc)
```

```python
print(classification_report(y_test, prediction))
```

<p><b>Precision</b> the ratio of correctly predicted positive observations to the total predicted positives.</p>

 For Iris-setosa, precision is <b>1.00</b>meaning that all instances predicted as Iris-setosa were actually Iris-setosa.
 
 For Iris-versicolor, the precision is <b>0.92</b>, indicating that <b>92%</b> of instances predicted as Iris-versicolor were correct.
 
 For Iris-virginica, the precision is <b>1.00</b>, meaning that all instances predicted as Iris-virginica were actually Iris-virginica.

<b>Recall (Sensitivity)</b> calculates the ratio of correctly predicted positive observations to all the actual positives. 

 For Iris-setosa, the recall is <b>1.00</b>, indicating that all actual Iris-setosa instances were correctly predicted.
 For Iris-versicolor, the recall is <b>1.00</b>, meaning that all actual Iris-versicolor instances were captured by the model.
 For Iris-virginica, the recall is  <b>0.92</b>, suggesting that <b>92%</b> of actual Iris-virginica instances were correctly predicted.


<b>F1-Score</b> The F1-score is the weighted average of precision and recall. It represents the balance between precision and recall. F1-score is high when both precision and recall are high.

  Iris-setosa has an F1-score of <b>1.00</b>, indicating a perfect balance of precision and recall.
  Iris-versicolor has an F1-score of <b>0.96</b>, showing a good balance between precision and recall.
  Iris-virginica also has an F1-score of <b>0.96</b>, indicating a good balance between precision and recall.



<b>Support</b> Support is the actual number of occurrences of the class in the specified dataset. For example, there are 7 instances of Iris-setosa, 11 instances of Iris-versicolor, and 12 instances of Iris-virginica in your test dataset.

<b>Accuracy</b> Accuracy is the ratio of correctly predicted instances to the total instances. the overall accuracy of the model on the test dataset is <b>97%</b>.

</b>macro avg</b> row provides the average scores across all classes

<b>weighted avg</b> row provides the weighted average of precision, recall, and F1-score, where each class's score is weighted by its support. 



**the model seems to be performing very well with high precision, recall, and F1-scores.**

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=121)
accs = []
```

```python
for train_i, test_i in kfold.split(X,y):
    X_train_fold, X_test_fold = X.iloc[train_i], X.iloc[test_i]
    y_train_fold, y_test_fold = y.iloc[train_i], y.iloc[test_i]
    
    clf.fit(X_train_fold, y_train_fold)
    fold_predict = clf.predict(X_test_fold)
    
    fold_acc = accuracy_score(y_test_fold, fold_predict)
    accs.append(fold_acc)
```

```python
accs
```

```python
## Mean and Standard Deviation
mean_acc = sum(accs)/len(accs)
print(f'Mean Accuracy   :{mean_acc:.2f}')
print(f"Std of Accuarcy :{(sum([(x-mean_acc)**2 for x in accs])/len(accs))**0.5:.2f}")
```
